Conceptos de POO (Programación Orientada a Objetos) en Python



###El prompt de IA para Gemini
lenguaje python, crear clase de animal con los atributos "nombre", "raza" crear el constructor animal, las funciones ladrar(), corres(), crear objetos pitbull, chihuahua, nivel principianre, no es interactivo

```python
class Animal:
    # Constructor de la clase
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función para ladrar
    def ladrar(self):
        print(f"{self.nombre} está ladrando: ¡Guau, guau!🙄")

    # Función para correr
    def correr(self):
        print(f"{self.nombre} de raza {self.raza} está corriendo rápidamente.🤦‍♀️")
    def Morder(self):
        print(f"{self.nombre}esta mordiendo a la vecina👩")

# Creación de los objetos (instancias de la clase)
pitbull = Animal("Rocky", "Pitbull")
chihuahua = Animal("Chispita", "Chihuahua")
# Creamos el objetos perro callejero
perrocallejero= Animal("Solovino", "Firulais")

# Uso de las funciones con los objetos creados
pitbull.ladrar()
pitbull.correr()

print("-" * 20)

chihuahua.ladrar()
chihuahua.correr()

print("-" * 20)

perrocallejero.ladrar()
perrocallejero.correr()

print("Programa creado por Iram Jimenez NL = 28")
```
